* Numărul total de rânduri
* Lungimea medie a cuvintelor
* Lista cuvintelor unice (apare o singură dată)
* Lista cuvintelor duplicate (apare de ≥ 2 ori)
* Numărul total de cuvinte unice
* Calculează statistici detaliate:
    - Numărul de vocale/consoane
    - Cuvântul cel mai lung
    - Raportul cuvinte unice / total cuvinte
    - Densitatea cuvintelor (cuvinte/caractere)

```python
def citeste_fisier_simplu(nume_fisier):
    """
    Citește întregul conținut al fișierului ca un singur string
    """
    try:
        with open(nume_fisier, 'r', encoding='utf-8') as f:
            continut = f.read()
        return continut
    except FileNotFoundError:
        print(f"Eroare: Fișierul {nume_fisier} nu există!")
        return None
    except Exception as e:
        print(f"Eroare la citire: {e}")
        return None

# Utilizare
text = citeste_fisier_simplu('Pupăza_din_tei.txt')
if text:
    print("Text citit:")
    print(text)
```
